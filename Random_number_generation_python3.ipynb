```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Random numbers will be generated using "numpy.random" module

```python
# Generate a random float between 0 & 1
x = np.random.rand() 
print(x)
# Generate a random integer between 0 & 50
x = np.random.randint(50)
print(x)
```

```python
# Generate an array of 5 random floats
x = np.random.rand(5)
print(x, x.shape)
# Generate an array of 5 random integers between 0 & 50
x = np.random.randint(50, size=(5))
print(x, x.shape)
# Generate a 2D array of random float
x = np.random.rand(5,3)
print(x, x.shape)
# Generate a 2D array of random integers between 0 & 50
x = np.random.randint(50, size=(5,3))
print(x, x.shape)
```

```python
# Pick a value randomly from an array
x = np.random.choice([1,2,3,4,5], size=(2,2))
print(x)
# Pick a value according to a predefined probability distribution from an array
x = np.random.choice([1,2,3,4,5], p=[0.1, 0.3, 0.3, 0.2, 0.1], size=10000)
sns.histplot(x, stat='probability') # Sum of histrogram heights = 1
plt.grid()
```

## Normal (Gaussian) distribution
$\Large P(x) = \frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{(x - \mu)^{2}}{2\sigma^{2}}}$\
Normal distribution is defined by mean=$\mu$ & std=$\sigma$

```python
x1 = np.random.normal(0, 1, size=(2000))
x2 = np.random.normal(0, 2, size=(2000))
# Plot probability distributions
sns.kdeplot(x1, fill=True, color='green', label=r'$\mathrm{\sigma=1}$')
sns.kdeplot(x2, fill=True, color='violet', label=r'$\mathrm{\sigma=2}$')
plt.legend()
```

## Binomial distribution
It is a discrete distribution
based on "Bernoulli trial", an experiment with two possible outcomes. One outcome (success) occurs with\
probability p & the other (failure) with probability (1-p).\
Q. What is the probability of getting k successes from n independent Bernoulli trials ?\
$\Large P(n, k) = ^{n}C_{k}p^{k}(1-p)^{n-k}$

```python
x = np.random.binomial(100, 0.5, size=10000)
#print(x) # Number of success over n trials
# Plot probability distributions
sns.kdeplot(x, fill=True, color='darkcyan')
```

## Uniform distribution
$\Large P(x) = \frac{1}{b-a}$

```python
x = np.random.uniform(-2, 2, 10000)
# Plot probability distributions
sns.kdeplot(x, fill=True, color='red')
```

## Poisson distribution
It is a discrete distribution. It estimates how many times an event can happen in a fixed interval of time.\
$\Large P(n;\lambda) = e^{-\lambda}\frac{\lambda^{n}}{n!}$ n = 0, 1, 2....\
Binomial distribution only has two outcomes, whereas poisson distribution can have unlimited possible outcomes.

```python
x1 = np.random.poisson(5, size=2000)
x2 = np.random.poisson(15, size=2000)
x3 = np.random.poisson(25, size=2000)
x4 = np.random.poisson(35, size=2000)


fig, ax = plt.subplots(figsize=(5,5))
sns.kdeplot(x1, ax=ax, fill=True, label=r'$\mathrm{\lambda=5}$')
sns.kdeplot(x2, ax=ax, fill=True, label=r'$\mathrm{\lambda=15}$')
sns.kdeplot(x3, ax=ax, fill=True, label=r'$\mathrm{\lambda=25}$')
sns.kdeplot(x4, ax=ax, fill=True, label=r'$\mathrm{\lambda=35}$')
plt.legend()
```
